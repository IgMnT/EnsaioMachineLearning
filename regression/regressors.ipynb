```python
import pandas as pd
import sklearn.metrics as mt
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets as ds
from sklearn import model_selection as ms
from sklearn import preprocessing as pp
from sklearn import metrics as mt
from sklearn import tree
from sklearn import svm
from sklearn.cluster import KMeans
from sklearn import neighbors
from sklearn import linear_model as lm
from sklearn import ensemble


```

```python
# Load dos datasets

x_train = pd.read_csv('X_training.csv')
x_test = pd.read_csv('X_test.csv')
y_train = pd.read_csv('y_training.csv')
y_test = pd.read_csv('y_test.csv')
x_validation = pd.read_csv('X_validation.csv')
y_validation = pd.read_csv('y_val.csv')

```

```python
# Linear Regression

# Metricas do train

reg = lm.LinearRegression().fit(x_train, y_train)
y_train_pred = reg.predict(x_train)
mt.r2_score(y_train, y_train_pred)
mt.mean_squared_error(y_train, y_train_pred)
mt.root_mean_squared_error(y_train, y_train_pred)
mt.mean_absolute_error(y_train, y_train_pred)
mt.mean_absolute_percentage_error(y_train, y_train_pred)
print("Train metrics")
print("R2: ", mt.r2_score(y_train, y_train_pred))
print("MSE: ", mt.mean_squared_error(y_train, y_train_pred))
print("RMSE: ", mt.root_mean_squared_error(y_train, y_train_pred))
print("MAE: ", mt.mean_absolute_error(y_train, y_train_pred))
print("MAPE: ", mt.mean_absolute_percentage_error(y_train, y_train_pred))

print("\n")  # Quebra de linha


# Metricas do validation
y_validation_pred = reg.predict(x_validation)
mt.r2_score(y_validation, y_validation_pred)
mt.mean_squared_error(y_validation, y_validation_pred)
mt.root_mean_squared_error(y_validation, y_validation_pred)
mt.mean_absolute_error(y_validation, y_validation_pred)
mt.mean_absolute_percentage_error(y_validation, y_validation_pred)
print("Validation metrics")
print("R2: ", mt.r2_score(y_validation, y_validation_pred))
print("MSE: ", mt.mean_squared_error(y_validation, y_validation_pred))
print("RMSE: ", mt.root_mean_squared_error(y_validation, y_validation_pred))
print("MAE: ", mt.mean_absolute_error(y_validation, y_validation_pred))
print("MAPE: ", mt.mean_absolute_percentage_error(y_validation, y_validation_pred))

print("\n")  # Quebra de linha


# Metricas do test
y_test_pred = reg.predict(x_test)
mt.r2_score(y_test, y_test_pred)
mt.mean_squared_error(y_test, y_test_pred)
mt.root_mean_squared_error(y_test, y_test_pred)
mt.mean_absolute_error(y_test, y_test_pred)
mt.mean_absolute_percentage_error(y_test, y_test_pred)
print("Test metrics")
print("R2: ", mt.r2_score(y_test, y_test_pred))
print("MSE: ", mt.mean_squared_error(y_test, y_test_pred))
print("RMSE: ", mt.root_mean_squared_error(y_test, y_test_pred))
print("MAE: ", mt.mean_absolute_error(y_test, y_test_pred))
print("MAPE: ", mt.mean_absolute_percentage_error(y_test, y_test_pred))



```
